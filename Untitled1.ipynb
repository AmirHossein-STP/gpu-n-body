```python
import matplotlib
matplotlib.use('dffr')

from pylab import *
import time


ion()

tstart = time.time()               # for profiling
x = arange(0,2*pi,0.01)            # x-array
line, = plot(x,sin(x))
for i in arange(1,2000):
    line.set_ydata(sin(x+i/10.0))  # update the data
    draw()                         # redraw the canvas

print ('FPS:' , 200/(time.time()-tstart))
```
